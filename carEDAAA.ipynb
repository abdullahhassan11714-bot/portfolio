```python
df['Car Sale Status'] = df['Car Sale Status'].map({'Sold' : 1 , 'Un Sold': 0})
df['Feedback'] = df['Feedback'].map({'Poor' : 1 ,'Average' : 2, 'Good' : 3 , 'Excellent' : 4})
```

```python
df['Days to Sell'] = df['Sold Date'] - df['Purchased Date']
df
```

```python
df.loc[df['Car Sale Status'] == 0 , 'Days to Sell'] = 0
df
```

```python
df['Profit quntity'] = df['Sold Price-$'] * (df['Margin-%'] /100)
df
```

```python
df['Location'].unique()
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(data=df , x='Location' , y='Purchased Price-$' , estimator=sum )
plt.xticks(rotation=45)
```

```python
top_makes = df.groupby("Car Name")["Purchased Price-$"].mean().nlargest(10).index
filtered_df = df[df["Car Name"].isin(top_makes)]

sns.barplot(data=filtered_df, x="Car Name", y="Purchased Price-$" , color='skyblue')
plt.xticks(rotation=45, ha='right')  # Rotate for better readability
plt.title("Top 10 Car Makes by Average Price")
plt.tight_layout()
```
